```python
import pandas as pd

def main():
    student_data = []
    invalid_data = []
    
    while True:
        student_name = input("Enter Student Name: ")
        student_class = input("Enter Class (10th or 12th): ")
        student_stream = input("Enter Stream (Science or Commerce): ")
        student_percentage = input("Enter Percentage: ")
        
#Data validation
        try:
            student_percentage = float(student_percentage)
        except ValueError:
            print("Percentage should be a number.")
            invalid_data.append([student_name, student_class, student_stream, student_percentage])
            student_percentage = None
        
        if not (student_name.isalpha() and student_class in ['10th', '12th']
                and student_stream in ['Science', 'Commerce'] and student_percentage is not None):
            print("Invalid data. This entry will be saved in the raw input file.")
            invalid_data.append([student_name, student_class, student_stream, student_percentage])
        else:
            student_data.append([student_name, student_class, student_stream, student_percentage])
        
        more_students = input("Do you want to enter details of another student? (yes/no): ")
        if more_students.lower() != "yes":
            break
    
#Saving both valid and invalid data to a CSV file
    df_student_data = pd.DataFrame(student_data, columns=["Student Name", "Class", "Stream", "Percentage"])
    
    if not df_student_data.empty:
        df_student_data.to_csv("studentinputdata36.csv", mode='a', header=False, index=False)
        print("Student input data saved to 'studentinputdata36.csv'.")
        
#Removing rows with missing percentage values   

    if invalid_data:

        invalid_data = [row for row in invalid_data if row[3] is not None]
        if invalid_data:
            df_invalid_data = pd.DataFrame(invalid_data, columns=["Student Name", "Class", "Stream", "Percentage"])
            df_invalid_data.to_csv("studentinputdata36.csv", mode='a', header=False, index=False)
            print("Invalid data saved to 'studentinputdata36.csv'.")
    
# Remove invalid rows from the valid data DataFrame
    df_student_data = df_student_data.dropna()
    
    if not df_student_data.empty:
#Adding the 'Grade' column to the valid data
        df_student_data["Grade"] = df_student_data["Percentage"].apply(assign_grade)
        
#Saving the processed data to a CSV file
        df_student_data.drop(columns=["Percentage"], inplace=True)
        df_student_data.to_csv("studentoutputdata36.csv", index=False)
        print("Valid data processed and saved to 'studentoutputdata36.csv'.")
    else:
        print("No valid data to process and save.")

def assign_grade(percentage):
    if percentage < 35:
        return "F"
    elif 35 <= percentage < 45:
        return "C"
    elif 45 <= percentage < 60:
        return "B"
    elif 60 <= percentage < 75:
        return "A"
    else:
        return "A+"

if __name__ == "__main__":
    main()
```
